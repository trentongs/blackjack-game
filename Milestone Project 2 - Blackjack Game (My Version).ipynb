```python
import random

ranks = ['two','three','four','five','six','seven','eight','nine','ten','King','Queen','Jack','Ace']
suits = ['Clubs','Diamonds','Hearts','Spades']
values = {'two':2,'three':3,'four':4,'five':5,'six':6,'seven':7,'eight':8,'nine':9,'ten':10,'King':10,
          'Queen':10,'Jack':10,'Ace':11}
```

```python
class Cards():
    
    def __init__(self,rank,suit):
        self.rank = rank
        self.suit = suit
        
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class Deck():
    
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Cards(rank,suit))
                
    def __str__(self):
        deck_comp = ''
        for card in self.deck:
            deck_comp += "\n" + card.__str__()
        return "The deck has: " + deck_comp
        
    def shuffle(self):
        random.shuffle(self.deck)
        
    def select(self):
        return self.deck.pop()
```

```python
class Action(Deck):  # Action class inherits Deck class
    
    def __init__(self):
        Deck.__init__(self) # Gonna create an instance of DrawDeck class when create instance of Action class
    
    def hit(self):
        print("\n\nHIT ME!")
        return self.select()
    
    def stay(self):
        print("\nI'll stay... I also like to live dangerously.")
```

```python
class Math():
    
    def __init__(self, nList):
        self.nList = nList
        self.value = 0
        self.ace = 0
            
    def add(self):
        for card in self.nList:
            self.value += values[card.rank]
            if card.rank == "Ace":
                self.ace += 1
                
    def adjustAce(self):
        while self.value > 21 and self.ace:
            self.value -= 10
            self.ace -= 1
        return self.value
```

```python
class Money():
    
    def __init__(self, total, bet, compSum, plyrSum):
        self.total = total
        self.bet = bet
        self.compSum = compSum
        self.plyrSum = plyrSum
        
    def autoWin(self):
        if self.plyrSum == 21:
            self.bet *= 2
            self.total += self.bet
            print(f"\n\nWINNER! You win {self.bet} chips")
            print("Total chips: ",self.total) 
            return 4
        else:
            return 1
        
    def revealWin(self):
        if self.compSum == 21:
            print(f"\n\nLOSER! You lose {self.bet} chips")
            print("Total chips: ",self.total)
            return 3
        else:
            return 1
    
    def winCheck(self):
        if self.plyrSum > self.compSum and self.plyrSum < 21:
            self.bet *= 2
            self.total += self.bet
            print(f"\n\nWINNER! You win {self.bet} chips")
            print("Total chips: ",self.total)
            return 4
        elif self.compSum > self.plyrSum and self.compSum < 21:
            print(f"\n\nLOSER! You lose {self.bet} chips")
            print("Total chips: ",self.total)
            return 3
        elif self.compSum == self.plyrSum < 21:
            self.total += self.bet
            print(f"\n\nTie! You win {self.bet} chips")
            print("Total chips: ",self.total)
            return 5
        else:
            pass

    def bustCheck(self):
        if self.compSum > 21:
            self.bet *= 2
            self.total += self.bet
            print(f"\n\nDealer BUSTS! You win {self.bet} chips")
            print("Total chips: ",self.total)   
            return 4
        elif self.plyrSum > 21:
            print(f"\n\nPlayer BUSTS! You lose {self.bet} chips")
            print("Total chips: ",self.total)
            return 3
        else:
            return 1
        
    def brokeCheck(self):
        if (self.total <= 0):
            print("You're all out of chips! Thanks for playing!")
            return 6
        else:
            pass
```

```python
class Update():
    
    def __init__(self, compList, playerList):
        self.compList = compList
        self.playerList = playerList
    
    def individual(self):
        print("\nDealer: ", end=" ")
        for i in self.compList:
            print(i, end="     ")
        print("\nPlayer: ", end=" ")
        for j in self.playerList:
            print(j, end="     ")
```

```python
class Automate():
    def __init__(self, compList, playerList, sum1, sum2, pot, wager,deal):
        self.compList = compList
        self.playerList = playerList
        self.sum1 = sum1
        self.sum2 = sum2
        self.pot = pot
        self.wager = wager
        self.deal = deal
        
    def dMoves(self):
        plTurn = 1
        while plTurn == 1:
            if self.sum1 < 17:
                extraC = self.deal.hit()
                self.compList.append(extraC)
                pr = Update(self.compList, self.playerList)
                pr.individual()
                # Add card values together
                val1 = Math(self.compList)
                val2 = Math(self.playerList)
                val1.add()
                val2.add()
                self.sum1 = val1.adjustAce()
                self.sum2 = val2.adjustAce()
                print(f"\nSum of Dealer cards: ", self.sum1)
                print(f"Sum of Player cards: ", self.sum2)
                chk2 = Money(self.pot,self.wager,self.sum1,self.sum2)
                plTurn = chk2.revealWin()      
                if plTurn == 1:
                    plTurn = chk2.bustCheck()
                    if plTurn == 4 or plTurn == 5:
                        return plTurn
                    else:
                        pass
                else:
                    pass
            else:
                self.deal.stay()
                pr = Update(self.compList, self.playerList)
                pr.individual()
                plTurn = 3
                # Add card values together
                val1 = Math(self.compList)
                val2 = Math(self.playerList)
                val1.add()
                val2.add()
                self.sum1 = val1.adjustAce()
                self.sum2 = val2.adjustAce()
                print(f"\nSum of Dealer cards: ", self.sum1)
                print(f"Sum of Player cards: ", self.sum2)
                 # Amount won and total money left is outputted when Player wins
                final = Money(self.pot, self.wager, self.sum1, self.sum2)
                return final.winCheck()
```

```python
def setChips():
    while True:
        try:
            tChips = int(input("\nTotal chips: "))
        except:
            print("Sorry please provide an integer")
        else:
            if tChips <= 0:
                print("Invalid amount, Please enter a number that is greater than 0")
            else:
                return tChips
                break   
```

```python
def setBet(chips):
    while True:
        try:
            bet = int(input("\nBet amount: "))
        except:
            print("Sorry please provide an integer")
        else:
            if bet > chips:
                print("Invalid amount, You have: {} chips".format(chips))
            else:
                return bet
                break   
```
